```python

import pandas as pd
import numpy as np
```

```python
wiki_postal_canada = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
table = wiki_postal_canada[0]
table
```

```python
table2 = table[table.Borough != 'Not assigned']
table2
```

```python
table3 = table2.reset_index(drop=True)
table3
```

```python
table3.shape
```

```python
geocord = pd.read_csv('Geospatial_Coordinates.csv', sep=",")
geocord
```

```python
geocord2 = pd.DataFrame(data=geocord)
geocord2
```

```python
answer2 = pd.merge(table3, geocord2, on=['Postal Code'])
answer2
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(answer2['Borough'].unique()),
        answer2.shape[0]
    )
)
```

```python

from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

address = 'Toronto, ON'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geographical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
map_toronto
```

```python
for lat, lng, borough, neighborhood in zip(answer2['Latitude'], answer2['Longitude'], answer2['Borough'],answer2['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
toronto_data = answer2[answer2['Borough'].str.contains('Toronto')].reset_index(drop=True)
toronto_data
```

```python
map_toronto2 = folium.Map(location=[latitude, longitude], zoom_start=10)
for lat, lng, borough, neighborhood in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Borough'],toronto_data['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto2)  
    
map_toronto2
```
